```python
#import "toolbox"
%matplotlib inline

def fix_zip(x):
    if x=='OOS':
        x=0
    elif x=='':
        x=0
    elif type(x) == type(8.8):
        x=0
    return int(x)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
sns.set(style="darkgrid")
```

```python
#import EV Registration Data.  This is the only table that has County and Zip on it.
#After Importing, pair down the dataframe so it only shows Zip and County
vehicle_reg = pd.read_csv('Vehicle__Snowmobile__and_Boat_Registrations.csv')
vehicle_reg[:5]
```

```python
county_zip = vehicle_reg[['Zip','County']]
county_zip = county_zip.drop_duplicates()
county_zip[:5]
```

```python
#upload population by zipcode data, convert series to dataframe
zip_pop = pd.read_csv('zip_pop.csv',converters={'Zip Code':fix_zip})
zip_pop = zip_pop.groupby('Zip Code')['Population'].sum()
zip_pop[:4]
```

```python
df_zip_pop = zip_pop.to_frame().reset_index()
df_zip_pop = df_zip_pop.rename(columns={0:'Zip Code',1:'Population'})
df_zip_pop[:4]
```

```python
#merge to create one table with zip, pop, and county
zip_pop_co = pd.merge(county_zip, df_zip_pop, how='inner', left_on='Zip', right_on='Zip Code')
zip_pop_co
```

```python
#Check the population of zip code 14207 for accuracy in merging tables
df_zip_pop[df_zip_pop['Zip Code']==14207]
```

```python
#Create a new data frame with county and population 
county_pop = zip_pop_co[['County', 'Population']]
county_pop[:5]
```

```python
#use groupby to get the sum of the populations for each county
total_co_pop = county_pop.groupby(['County'], as_index=False).sum()
total_co_pop
```

```python
county = vehicle_reg[['Zip','County']]
county[:5]
```

```python
county['Number of Vehicles'] = ''
reg_county = county.groupby(['County'], as_index = False)
count_county = reg_county.agg({'Number of Vehicles':'count'}).sort_values(['County'], ascending= True)
count_county
```

```python
count_county_area['EV Density by Land Area'] = count_county_area['Number of Vehicles'] / count_county_area['2000 Land Area Square Miles']
count_county_area [:5]
```

```python
count_county_area['EV Density by Population'] = count_county_area['Number of Vehicles'] / count_county_area['2006 Population Estimate']
count_county_area[:5]
```
